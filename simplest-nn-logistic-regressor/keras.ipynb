```python
from keras.layers import Dense, Activation
from keras.models import Sequential
```

```python
model = Sequential()

# layer 1
model.add(Dense(3, input_dim=2)) # Hidden layer of size 3, input values are spatial coordinate => 2 dimension
model.add(Activation('tanh')) # Add tanh activation function for first layer
# Output layer
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.summary()

# Note Output Shape always have None for the first 'dimension' to accept any number of samples
```

```python
model.compile(optimizer='sgd',  # Use stochastic gradient descent
              loss='binary_crossentropy',
              metrics=['acc'])


import numpy as np
np.random.seed(1)

# Define input X and desired output y
X = np.array([[0, 1], [1, 0], [1, 1], [0, 1]])
y = np.array([[0, 1, 1, 0]]).T

history = model.fit(X,y,epochs=900)
```
